```python
import ctypes

def get_dpi():
    LOGPIXELSX = 88
    LOGPIXELSY = 90
    user32 = ctypes.windll.user32
    user32.SetProcessDPIAware()
    dc = user32.GetDC(0)
    dpi_x = ctypes.windll.gdi32.GetDeviceCaps(dc, LOGPIXELSX)
    dpi_y = ctypes.windll.gdi32.GetDeviceCaps(dc, LOGPIXELSY)
    user32.ReleaseDC(0, dc)
    return dpi_x, dpi_y

dpi_x, dpi_y = get_dpi()
print(f"Horizontal DPI: {dpi_x}, Vertical DPI: {dpi_y}")
```

```python
from AppKit import NSScreen
from Quartz import CGDisplayScreenSize

for i, screen in enumerate(NSScreen.screens(), 1):
    description = screen.deviceDescription()
    screen_number = description["NSScreenNumber"]
    # Get pixel dimensions
    width_px = int(screen.frame().size.width * screen.backingScaleFactor())
    height_px = int(screen.frame().size.height * screen.backingScaleFactor())
    # Get physical size in millimeters
    width_mm, height_mm = CGDisplayScreenSize(screen_number)
    # Convert mm to inches
    width_in = width_mm / 25.4
    height_in = height_mm / 25.4
    # Calculate DPI
    dpi_x = width_px / width_in
    dpi_y = height_px / height_in
    print(f"Display #{i}: {dpi_x:.1f} x {dpi_y:.1f} DPI")
```
